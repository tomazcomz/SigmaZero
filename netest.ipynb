```python
x,yp,yv=optimizar.create_train_set(game)
```

```python
yv
```

```python
len(x)
```

```python
rede=Neura(game,name='acacio')
rede.compilar()
```

```python
ARGS = {
    'cpuct': 1.5,
    'num_searches': 100
}
alpha=MCTS.MCTS(game,ARGS,rede)
```
